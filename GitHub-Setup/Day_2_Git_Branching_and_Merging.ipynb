## Goals
 1. Understand branches
 2. create and switch between them
 3. merge code
 4. resolve conflicts
 5. understand PRs (Pull Requests - very important for interviews).

### Branch

It is a separate line of development that lets you work on feature without affecting main code. It makes collaboration safe and organised.

main branch = Production code

feature branch = new work

### Branch Commands

1. Creation of new branch

    git branch feature-login
    
    
2. Switch to a branch

    git checkout feature-login
    
    
3. Create AND Switch (shortcut)

    git checkout -b feature-login
    
    
4. List all branches

    git branch
    
    
5. Delete a branch (after merge)

    git branch -d feature-login
    

### Git Workflow with Branching

1. Checkout main


2. Pull the latest code

    git checkout main
    git pull origin main
    
    
3. Create feature branch and checkout feature branch

    git branch feature-login
    git checkout -b feature-login
    
    
4. Make changes


5. Add and Commit

    git add .
    git commit -m "Add Feature login"
    
    
6. Push your branch

    git push origin feature-login

### Merge Branches

1. Switch to the branch you want to merge into.

    git checkout main
    
    
2. Merge your feature branch into main.

    git merge feature-login
    

### Merge Conflicts

Merge Conflicts happen when,

1. Two branches edit the same line in a file

2. Git does not know which version to keep.


Solution

1. Manually edit the file

2. Keep the correct code

3. Add and commit the file

    git add <file-name>
    
    git commit -m "Resolve Merge Conflicts"

### Pull Requests

It is a request to merge your branch to another branch.
It includes:
1. Summary of Changes
2. Code Reviews
3. Comments
4. Checks(like CI/CD)
