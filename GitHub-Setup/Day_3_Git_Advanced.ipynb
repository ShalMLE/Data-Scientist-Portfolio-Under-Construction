# Goal: 

- Understand how to view history
- compare changes, 
- temporarily save changes,
- undo mistakes safely.

## Git log - View Commit History

#### Basic log
git log

#### Short log
git log --oneline

#### log with graph
git log --oneline --graph --decorate --all

### Use Cases

1. Checking past versions
2. Checking specific Commit
3. Understanding where Branches Diverged

## Git diff - Compare Changes

#### Comparing working directory with Staged files
git diff

#### Comparing Staged files and last Commit
git diff --staged

#### Comparing two branches
git diff master feature-branch

#### Comparing two Commits
git diff <commit1> <commit2>
    
Eg. git diff eda-v1.ipynb eda-v2.ipynb

## Git Stash - Save file without Commiting

Purpose - Store your uncomitted changes temporarily, so that you can switch branches safely.

#### Stash Changes
git stash

#### See all Stashes
git stash list

#### Apply Stash/Keep Stash
git stash apply

#### Apply and Remove Stash
git stash pop

#### Drop Stash
git stash drop stash@{0} 


### Use case

You started EDA work but you need to pull latest changes so you stash the changes temporarily.

## Git Revert - Safest way to revert the changes

To undo a commit by creating a new commiting that revert the changes.

git revert <commit-id>
    
Safer - It does not rewrite commit history,suitable for shared branches like main.

# Git Reset - Powerful but Risky

### Soft Reset 

Moves HEAD back but files remain Staged.

git reset --soft {commit-id}
    

### Mixed Reset
    
Moves HEAD back but keep changes in the working directory.
    
git reset {commit-id}
    

### Hard Reset - To be used Only on local branches
    
Delete all changes which cannot be recovered.
    
git reset --hard {commit-id}

## Git Restore - New Command

### Restore working Directory file
git restore {file-name}

### Restore Staged file
git restore --staged {file-name}

## Git Clean - Remove Untracked File

Very Useful in ML/DS where notebooks generate unwanted files.

#### Preview
git clean -n 

#### Delete
git clean -f
