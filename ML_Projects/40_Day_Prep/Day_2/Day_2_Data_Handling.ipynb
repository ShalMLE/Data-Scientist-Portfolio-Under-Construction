# Pandas Basics

What is Pandas?

Used for:

tables

data cleaning

EDA

grouping

merging datasets

handling missing data

```python
#DataFrame Creation
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Age": [23, 25, 31],
    "City": ["NY", "LA", "TX"]
}

df = pd.DataFrame(data)
df
```

```python
#Loading the data
data = pd.read_csv('file.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#Selecting Columns
data['Age']
```

```python
data[['Age','Salary']]
```

```python
#Filtering Rows
data[data['Age'] > 30]
```

```python
data[(data['Age'] > 30) & (data['Salary'] > 50000)]
```

```python
#Adding & Removing Columns
data['Age_2'] = data['Age'] * 2
```

```python
data.head()
```

```python
data.drop('Age_2',axis=1)
```

```python
#Handling Missing Values
data.isnull().sum()
```

```python
data.fillna(data.mean())
```

```python
data.dropna()
```

```python
#Groupby
data.groupby('City')['Age'].mean()
```

```python
data.groupby('City').agg({'Age':['mean','max','min']}).T
```

```python
#Sorting
data.sort_values('Age',ascending=False)
```

```python
#Merging
pd.merge(data,data,on='Name',how='inner')
```

### Pandas-Tasks

Task 1:
Create a DataFrame with columns:

Product

Price

Category

Add 5 rows.

```python
data1 = {
    "Product":['P1', 'P2', 'P3', 'P4', 'P5'],
    "Price":[30, 35, 15, 12, 20],
    "Category":['Induction', 'Dryer', 'Mixer', 'Induction', 'Dryer']
}

df = pd.DataFrame(data1)
df.head()
```

Task 2:
Filter all products whose price > $20.

```python
df[df['Price'] > 20]
```

Task 3:
Group by Category â†’ find average price.

```python
df.groupby('Category')['Price'].mean()
```

Task 4:
Add a new column â€œDiscounted Priceâ€ with 10% discount.

```python
df['Discounted Price'] = df['Price'] - df['Price'] * .10
```

```python
df
```

Task 5:
Create two small DataFrames and practice merge (inner + left).

```python
data2 = {
    "Product":['P6', 'P7', 'P8', 'P9', 'P10'],
    "Price":[70, 45, 25, 22, 15],
    "Category":['Induction', 'Dryer', 'Mixer', 'Induction', 'Dryer']
}

df2 = pd.DataFrame(data2)
df2.head()
```

```python
pd.merge(df,df2,on='Category',how='inner')
```

```python
pd.merge(df,df2,on='Category',how='left')
```
